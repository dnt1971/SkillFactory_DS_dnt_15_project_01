1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
#ваш код здесь

#######################################################
#
# Вычленяет из первых 2х слов входной строки уровень образования и возвращает 1 из 4х вариантов
#  "высшее", "неоконченное высшее", "среднее специальное", "среднее"
#
# Вход: строка, первые 2 слова - уровень образования
# Выход: Одна из 4х строк: "высшее", "неоконченное высшее", "среднее специальное", "среднее
# Пример:
# 
#    "Высшее образование 2002  Саратовский государственный университет им. Н.Г. Чернышевского, Саратов Филологический, филология (лингвистика), Красный диплом, специализация: филолог (лингвист), доп. специализация: информационные технологии в филологии"
#    ==> "высшее"
#######################################################
def cvt_obr( s_msg ):
    lev = { 
        "высшее образование" : "высшее", 
        "неоконченное высшее": "неоконченное высшее", 
        "среднее специальное": "среднее специальное", 
        "среднее образование": "среднее"
    }
    return lev.get( ' '.join( s_msg.lower().split()[:2] ) )

# Добавляем новый признак
hh_data["Образование"] = hh_data["Образование и ВУЗ"].apply( lambda x: cvt_obr(x) ) 

print( "Cоискателей, имеющих средний уровень образования : {}".format( hh_data["Образование"].value_counts()['среднее'] ) )
#559
print( "Количество уникальных значений признака образования : {}".format( hh_data["Образование"].nunique() ) )
# 4

# Удаляем признак
hh_data.drop( "Образование и ВУЗ", axis = 1, inplace = True )

# Проверим уникальность
# hh_data["Образование"].unique()
# array(['неоконченное высшее', 'высшее', 'среднее специальное', 'среднее'], dtype=object)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь

######################################################
#
# Вход: строка с годами и месяцами формата: Опыт работы [<N [год, года, лет]> ][<M [месяцев, месяца, месяц]>] ...
#       разделитель - пробел
# Выход: Число, количество месяцев
#    1) Если данных нет, или на входе задана строка "не указано", то возвращаем NaN
#    2) Учитываем первые вхождения фраз "год","года","лет" и "месяц","месяца","месяцев"
#
# Примеры:
#   Опыт работы 16 лет 10 месяцев  Август 2010 — по настоящее время 8 лет 10 месяцев МАОУ ""СОШ № 1 г.Немана"" Системный администратор Обслуживание ПК,установка ПО, ремонт, периферийной техники, Интернет локальная сеть. Ведение Электронного журнала, сайта организации.  Август 2002 — Август  2010 8 лет 1 месяц ТС ""ВЕСТЕР-ИНФО"" Старший продавец, директор отдела Продажи компьютерной техники"
#   ==> Опыт работы 16 лет 10 месяцев ==> 202
#
######################################################
def cvt_opyt( s_msg ):
    if pd.isna( s_msg ):
        return np.nan    
    
    # Приводим к нижнему регистру и убираем пробелы
    s_msg = str( s_msg ).strip().lower() 
    if s_msg == "не указано":
        return np.nan      
      
    arr_words_year = ['года','год','лет']
    arr_words_month = ['месяцев','месяца','месяц']
    tmp_work_month = 0
    tmp_work_year = 0
    
    lst = s_msg.split()
    # Соласно формату, слова 3-6 подвергаются анализу, (2-3),[(4-5)] 
    if len( lst ) >= 4: 
        # Примеры слов 3 и 4: "3 года" или "11 месяцев"
        if lst[3] in (arr_words_year):
            tmp_work_year = int(lst[2])
        elif lst[3] in (arr_words_month):
            tmp_work_month = int(lst[2])        
        
     
    if len( lst ) >= 6:
        # Примеры слов 5 и 6: "3 года" или "11 месяцев" 
        # Считаем только 1е вхождение слов (те ошибки типа "3 мес 11 месяцев" считываем только 1ю фразу)
        if tmp_work_year == 0 and lst[5] in (arr_words_year):
            tmp_work_year = int(lst[4])
        elif tmp_work_month == 0 and lst[5] in (arr_words_month):
            tmp_work_month = int(lst[4])
    
    return tmp_work_year * 12 + tmp_work_month

hh_data["Опыт работы (месяц)"] = hh_data["Опыт работы"].apply( lambda x: cvt_opyt( x ) )
print( "Медианный опыт работы (в месяцах) : {}".format( hh_data['Опыт работы (месяц)'].median() ) )
# 100
hh_data.drop( "Опыт работы", axis = 1, inplace = True )
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь

#######################################################
#
# Получает из входной строки список строк 
# Вход: Строка формата <Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>
# Выход: Список из 2х или 3х элементов [Страна, готовность к переезду, , готовность к командировкам]
# Особенности:
#   1) Выкусываем строки в открывающей и закрывающей круглых скобках (с самими скабками). 
#   Учитываем возможную вложенность скобок. Если открывающая скобка не закрыта - 
#   выкусываем всю эту часть до конца входной строки
#   2) Удаляем фразу - метро 
#
# Примеры: 
#   "Советск (Калининградская область) , не готов к переезду , не готов к командировкам" ==> ["Советск", "не готов к переезду", "не готов к командировкам"]
#   "Москва , не готова к переезду , готова к командировкам" ==> ["Москва", "не готова к переезду", "готова к командировкам"]
# 
#######################################################
def cvt_town_move( s_msg ):
    
    # Рекурсивно очищает от содержимоно круглых скобок входящую строку (и от самих скобок)
    # Вход: строка
    # Выход: очищенная от круглых скобок строка. Учитывается возсожность незакрытых скобок
    # Примеры:
    #   12(345)67 ==> 1267
    #   12(3(4)5)67 ==> 1267
    #   12(34567 ==> 12
    #   12)34567 ==> 12)34567
    def clear_backet( s_msg ):
        ind1 = s_msg.rfind( '(' ) # Самая правая открывающая скобка
        if ind1 < 0:
            return s_msg
        ind2 = s_msg[ind1 + 1:].find( ')' )  # Самая близкая слева закрывающая скобка
        
        if ind2 < 0: # Если закрывающей сковки нет - откусываем, начиная с открывающей скобки и запускаем процедуру к оставшейся строке
            return clear_backet( s_msg[:ind1] )
        return clear_backet( s_msg[:ind1] + s_msg[ind1 + 1 + ind2 + 1:] )  # Откусываем интервал и запускаем процедуру к оставшейся строке

    if pd.isna( s_msg ):
        return np.nan
    
    s_msg = str( s_msg )
        
    # Готовим к очистке от скобок (убрав пробел перед открывающей скобкой)
    s_msg = s_msg.replace( ' (', '(' )
    # Чистим от скобок с содержимым
    s_msg = clear_backet( s_msg )
    # Разбиваем на поля
    lst = s_msg.split( ' , ' )
    # Согласно формату - выкидываем элемент с метро
    if len( lst ) > 1:
        if lst[1].find( 'м. ' ) >= 0:
            lst.pop(1)
    # Возвращаем список первых 3х слов
    return lst[:3]

#######################################################
#
# Извлекает согласно формату название города из входной строки
# Преобразует полученный годор по правилу
#   - "Москва", "Санкт-Петербург" не преобразовываем
#   - преобразовать города из списка ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
#     в "город-миллионник"
#   - иначе возвращаем "другие"
# 
# Примеры:
#   "Озерск (Челябинская область) , не готов к переезду , не готов к командировкам" ==> "другие"
#   "Москва ,  м. Славянский бульвар , готов к переезду , готов к командировкам" ==> "Москва"
#
#######################################################
def cvt_town( s_msg ):
    if pd.isna( s_msg ):
        return "другие"
    
    # 1е слово согласно формату - название города. 
    # Тут же учтем возможность заполнения метро
    # Извлечем 1е слово, приведем его в нижний регистр и без пробелов слева и права
    tst_town = cvt_town_move( str( s_msg ) )[ 0 ].strip().lower()
    towns = {
        "москва": "Москва",
        "санкт-петербург": "Санкт-Петербург",
    }
    
    # Проверим на большие города
    find_town = towns.get( tst_town )
    if find_town is not None:
        return find_town 
    
    # Проверим на миллионники
    million_cities = list( map( lambda x: x.lower(), ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград'] ) )
    if tst_town in million_cities:
        return "город-миллионник"
    
    return "другие"

#######################################################
#
# Извлекает из входной строки признак "Готовность к переезду"и возвращает булево True или False
# Разделитель фраз " , "
# Использует словарь (полученный из массива данных), приведенный к нижнему регистру
#
#######################################################
def cvt_ready_move( s_msg ):
    # При аномальных данных считаем, что не готов к переезду
    if pd.isna( s_msg ):
        return False
        
    lst_words = cvt_town_move( str( s_msg ) )
    if len( lst_words ) < 2:
        return False
    
    # 2я фраза согласно формату - признак "Готовность к переезду"
    # приводим к нижнему регистку и убираем пробелы слева и справа
    word_2 = cvt_town_move( str( s_msg ) )[ 1 ].strip().lower()

    # Остались только такие варианты
    arr_vars = { 
        'не готов к переезду': False,
        'готов к переезду': True,
        'не готова к переезду': False,
        'готова к переезду': True,
        'хочу переехать': True,
        'готов к переезду ,': True
    }
    return arr_vars.get( word_2 )

#######################################################
#
# Извлекает из входной строки 3й паакет фраз - признак "Готовность к командировкам"и возвращает булево True или False
# Разделитель фраз " , "
# Использует словарь (полученный из массива данных), приведенный к нижнему регистру
#
#######################################################
def cvt_ready_remote_work( s_msg ):
    # При аномальных данных считаем, что не готов к переезду
    if pd.isna( s_msg ):
        return False
        
    lst_words = cvt_town_move( str( s_msg ) )
    if len( lst_words ) < 3:
        return False
    
    arr_vars = { 
        'готова к редким командировкам': True, 
        'готов к редким командировкам': True,
        'не готова к командировкам': False, 
        'не готов к командировкам': False, 
        'готова к командировкам': True, 
        'готов к командировкам': True, 
        'готов к командир': True,
        'готов к команд': True,
        'готов к': True,
        'г': True
    }
    return arr_vars.get( lst_words[ 2 ] )
    
# Добавляем признаки
hh_data["Город"] = hh_data["Город, переезд, командировки"].apply( lambda x: cvt_town( x ) ) 
hh_data["Готовность к переезду"] = hh_data["Город, переезд, командировки"].apply( lambda x: cvt_ready_move( x ) )
hh_data["Готовность к командировкам"] = hh_data["Город, переезд, командировки"].apply( lambda x: cvt_ready_remote_work( x ) )

################ Проверим данные
print( hh_data[[ "Город", "Готовность к переезду", "Готовность к командировкам" ]].info() )
# RangeIndex: 44744 entries, 0 to 44743
# Data columns (total 3 columns):
#      Column                      Non-Null Count  Dtype 
# ---  ------                      --------------  ----- 
#  0   Город                       44744 non-null  object
#  1   Готовность к переезду       44744 non-null  bool  
#  2   Готовность к командировкам  44744 non-null  bool  
print( hh_data[ "Город" ].unique() )
# ['другие' 'Москва' 'город-миллионник' 'Санкт-Петербург']

print( "Соискатели, живущие в Санкт-Петербурге : {}".format( round( hh_data[hh_data["Город"] == "Санкт-Петербург"].shape[0] * 100 / hh_data.shape[0] ) ) )
#11%
print( "Готовы одновременно и к переездам, и к командировкам : {}".format( round( hh_data[ hh_data["Готовность к переезду"] & hh_data["Готовность к командировкам"]].shape[0] * 100 / hh_data.shape[0] ) ) )
#32%

hh_data.drop( "Город, переезд, командировки", axis = True, inplace = True )
```

На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>

```python
#ваш код здесь

# Готовим словарь преобразования
works = {
    'Занятость': ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка'],
    'График': ['полный день', 'сменный график', 'гибкий график' ,'удаленная работа', 'вахтовый метод']
}

# И по совпадению создаем булев признак. 
for work_cols in works:
    for work_col in works[work_cols]:
        hh_data[work_col] = hh_data[work_cols].apply( lambda x: work_col in x.split( ', ' ) )

################ Проверим данные
print( hh_data[[ 'полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка', 'полный день', 'сменный график', 'гибкий график' ,'удаленная работа', 'вахтовый метод' ]].info() )


print( "Ищут проектную работу и волонтерство : {}".format( hh_data[hh_data['проектная работа'] & hh_data['волонтерство']].shape[0] ) )
#436
print( "Хотят работать вахтовым методом и с гибким графиком  : {}".format( hh_data[hh_data['вахтовый метод'] & hh_data['гибкий график']].shape[0] ) )
#2311

hh_data.drop( ['Занятость','График'], axis = 1, inplace = True )
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.

Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

# Прочитаем файл с курсами валют
money_curs = pd.read_csv('data/ExchangeRates.csv', delimiter = ',' )
# Превратим дату курса из текстового формата в тип даты 
money_curs["date"] = money_curs["date"].apply( lambda x: pd.to_datetime( x, dayfirst = True, yearfirst = False) )

# Создадим таблицу преобразования в iso из формата текстового файла
df_money = pd.DataFrame(
    data = {
        #"money_csv" : ["грн","USD","EUR","белруб","KGS","сум","AZN","KZT","руб"],
        "money_iso" : ['UAH','USD','EUR','BYN','KGS','UZS','AZN',"KZT","RUB"],
        "Пропорция" : [10,1,1,1,10,10000,1,100,1]
    }
    , index = ["грн","USD","EUR","белруб","KGS","сум","AZN","KZT","руб"]
)

# Превратим дату подачи из текстового формата в тип даты
hh_data["Обновление резюме"] = hh_data["Обновление резюме"].apply( lambda x: pd.to_datetime(x, dayfirst = True, yearfirst = False) ).dt.normalize()
# Извлечем признак суммы из композитного признака ЗП ("29000 руб." ==> 29000)
hh_data["ЗП_1"] = hh_data["ЗП"].apply( lambda x: x.split()[0] ).astype(int)
# Извлечем типа вылюты из композитного признака ЗП  ("29000 руб." ==> "руб")
hh_data["ЗП_2"] = hh_data["ЗП"].apply( lambda x: x.split()[1].replace( '.', '' ) )

# Добавим из df_money название валют в ISO формате
hh_data = hh_data.merge(
    df_money,
    how = 'left',
    left_on = "ЗП_2",
    right_index = True
)

# Склеиваем с датафреймом курсов валют по полям 'Обновление резюме', 'money_iso'
money_curs['Обновление резюме'] = money_curs['date']
money_curs['money_iso'] = money_curs['currency']

hh_data = hh_data.merge(
    money_curs,
    on = ['Обновление резюме','money_iso'],
    how = 'left',
)
money_curs.drop( ['Обновление резюме','money_iso'], axis = 1 )

# рубли не попали в курс - зададим вручную
hh_data.loc[hh_data["ЗП_2"] =="руб","close"] = 1

# Искомую сумму занесем в признак
hh_data["ЗП (руб)"] = ( hh_data["ЗП_1"] * hh_data["close"] / hh_data["Пропорция"]).round()

print( "Желаемая медианная заработная плата соискателей : {}".format( hh_data["ЗП (руб)"].median() ) )
#59019.0 ==> 59 (тыс.руб)

# Избавляемся от технических полей
hh_data.drop( ['ЗП','ЗП_1','ЗП_2','currency','money_iso','close','proportion','per','date','time','Пропорция','vol'], axis = 1, inplace = True )
```

5. Постройте диаграмму, которая показывает распределение желаемой заработной платы (**"ЗП (руб)"**) в зависимости от города (**"Город"**). Используйте для диаграммы данные о резюме, где желая заработная плата меньше 1 млн рублей.
*Сделайте выводы по полученной диаграмме: как соотносятся медианные уровни желаемой заработной платы и их размах в городах? Как вы считаете, важен ли признак города при прогнозировании заработной платы?*

```python
# ваш код здесь

#############################################
# 
# Вход: Датафрейм, признак группировки и признак, по которому вычисляются статистические функции
# Выход: Датафрейм с вычисленными статистическими функциями в зависимости от класса городов
#
#############################################
def df_analyse( data, cols_grp, col_value ):
    
    def q25(x):
        return x.quantile(0.25)

    def q75(x):
        return x.quantile(0.75)

    def iqr(x):
        return x.quantile(0.75) - x.quantile(0.25)

    def median_iqr(x):
        return (x.median() * 100 / (x.quantile(0.75) - x.quantile(0.25))).round().astype(int)
    
    tmp_df = data.copy()
    df_grp = tmp_df.groupby( cols_grp, as_index = True)[col_value].agg( ['median','std', q25, q75, iqr, median_iqr])
    
    return df_grp    

# Берем нужные признаки с ЗП < 1000000
tmp_df = hh_data[ hh_data['ЗП (руб)'] < 1000000 ][['ЗП (руб)','Город']].copy()
# Готовим датафрейм со статистикой ЗП по признаку "Город"
tmp_df_grp = df_analyse( tmp_df, 'Город', 'ЗП (руб)' ).sort_values( by = 'median_iqr' ) #.reset_index() #.sort_values( by = 'median_iqr' )

# Размечаем две координатные плоскости в 1 строку
fig = make_subplots(rows = 1, cols = 2 )

# Создаем графические объекты
fig1 = go.Box( 
    #x = tmp_df['ЗП (руб)'], 
    y = tmp_df['ЗП (руб)'], 
    x = tmp_df['Город'], 
    #name = '',   
)
fig2 = go.Bar(
    x = tmp_df_grp.index, 
    y = tmp_df_grp['median_iqr'], 
    orientation = 'v',
    text = tmp_df_grp['median_iqr']
    #name = "",
)
# И кладем в ячейки графической области
fig.add_trace( fig1, row = 1, col = 1 ) # Гистограмма, Количество = f( Возраст )
fig.add_trace( fig2, row = 1, col = 2 ) # Коробка с усами, Статистика и выбросы = f( Возраст )

fig.update_layout(
    title = "Исследуем медианную зарплату в зависимости от класса городов",
    title_x = 0.5, # Центрируем по оси x заголовок
    showlegend = False, # Отключаем
)

fig['layout'].update( width = 1000, height = 500)

# Подписываем оси графиков на обоих рисунках
fig.update_yaxes( title = 'ЗП (руб)', col = 1, row = 1)
fig.update_yaxes( title = 'Медианная ЗП / межквартильный размах, %', col = 2, row = 1)
fig.data[1].marker.color = ('green','lightgreen','deeppink','red')

fig['layout'].update( width = 800, height = 600)
# Выводим графику
fig.show()
```

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
# ваш код здесь

##########################################################
#
# функция возвращать выбросы и очищенные от них данные методом 3х сигм
# Вход:
#    data - Путь к csv-файлу или датафрейм
#    feature - Анализируемый нав выбросы признак
#    log_scale - сжатие  (путем логарифмирования) величин признака. По умолчанию - выключен
#    left, right - количество стандартных отклонений слева и справа от мат.ожидания распределения (по умолчанию = 3)
# Выход:
#    outliers - датафрейм выбросов
#    cleaned - датафрейм чистых данных
#
##########################################################
def outliers_z_score_mod( data, feature, log_scale = False, left = 3, right = 3):
    if isinstance(data, str):  # Проверка, если передан путь к файлу
        data = pd.read_csv(data)
               
    # Получаеи маспределение 
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
        
    mu = x.mean()   # Матожидание
    sigma = x.std() # 1 сигма = 1 стандартное отклонение
    lower_bound = mu - left * sigma  # Настроим левую границу
    upper_bound = mu + right * sigma # И правую
    
    # Получаем выбросфы и чистые данные
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score_mod(hh_data, 'Возраст', log_scale = True, left = 3, right = 4 )
print( f'Число выбросов по методу z-отклонения: {outliers.shape[0]}' )
print( f'Результирующее число записей: {cleaned.shape[0]}' )

display(outliers["Возраст"])
#31137     15
#32950     15
#33654    100
```
